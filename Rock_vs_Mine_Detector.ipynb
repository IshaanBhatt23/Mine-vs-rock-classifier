```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
sonar_data = pd.read_csv('/content/drive/MyDrive/dataset/sonar.all-data.csv')
```

```python
sonar_data.dropna(inplace=True)
```

```python
x = sonar_data.drop(columns='Label', axis=1)
y = sonar_data['Label']
```

```python
print(x.shape)
```

```python
print(y.shape)
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_encoded, test_size=0.2, random_state=42)
```

```python
model = Sequential([ Dense(32, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid') ])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=50, batch_size=8, validation_split=0.2)
```
